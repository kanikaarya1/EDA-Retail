### A lot of outliers are present in the dataset so we need to remove them to remove false predictions.Hence removing outliers using Interquartile range method.

## Step16: Applying IQR method to remove outliers present in the dataset.

```python

numeric_columns = df1.select_dtypes(include=['number']).columns

df1[numeric_columns] = df1[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Calculate quartiles and IQR for each numeric column
Q1 = df1[numeric_columns].quantile(0.25)
Q3 = df1[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

print(IQR)
```

## Step17: Reshaping dataset after taking all the outliers out.

```python

numeric_columns = df1.select_dtypes(include=['number']).columns

# Convert only numeric columns to numeric type, handling non-numeric values
df1[numeric_columns] = df1[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Calculate quartiles and IQR for each numeric column
Q1 = df1[numeric_columns].quantile(0.25)
Q3 = df1[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Filter out rows with outliers
df2 = df1[~((df1[numeric_columns] < (Q1 - 1.5 * IQR)) | (df1[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]

print('The modified shape of the dataset is:')
print(df2.shape)
```
